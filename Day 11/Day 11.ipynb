__Importing required package for correlation analysis__

```python
from scipy.stats import pearsonr
```

__Checking Correlation between Attrition and Distance from Home__

```python
stats, p=pearsonr(dataset['Attrition'], dataset['DistanceFromHome'])
print(stats, p)
```

 - Here p value is greater then 0.05
 - So Null hypothesis is accepted
 - There is no signinficant correlation between Attrition and Distance from home
